```python
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import mutual_info_regression
import pandas as pd

# Function to calculate cross-validation accuracy score
def score(model, X, y, cv=5):
    """
    Evaluates the model using cross-validation and returns the mean accuracy score.

    Parameters:
    - model: The machine learning model
    - X: Feature matrix (DataFrame)
    - y: Target variable (Series)
    - cv: Number of cross-validation folds (default=5)

    Returns:
    - Mean accuracy score
    """
    return cross_val_score(model, X, y, cv=cv, scoring='accuracy').mean()

# Function to compute Mutual Information (MI) scores for feature selection
def get_mutual_info_scores(X, y):
    """
    Computes the mutual information scores between features and the target variable.

    Parameters:
    - X: Feature matrix (DataFrame)
    - y: Target variable (Series)

    Returns:
    - Pandas Series with MI scores sorted in ascending order
    """
    mi_scores = mutual_info_regression(X, y, random_state=0)
    mi_scores = pd.Series(mi_scores, index=X.columns)
    return mi_scores.sort_values(ascending=True)
```

```python
#Train & Test Set
import pandas as pd
from sklearn.model_selection import train_test_split # Import the train_test_split function

X= df.iloc[: , 1:-1]
#y = upsampled_df['Churn']
y= df['target']

train_x,test_x,train_y,test_y = train_test_split(X,y,random_state=42)
print("\n--Training data samples--")
print(train_x.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the model
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(train_x, train_y)

# Predictions
log_reg_pred = log_reg.predict(test_x)

# Evaluation
log_reg_acc = accuracy_score(test_y, log_reg_pred)
print(f"Logistic Regression Accuracy: {log_reg_acc:.4f}")
print("\nClassification Report:\n", classification_report(test_y, log_reg_pred))
print("\nConfusion Matrix:\n", confusion_matrix(test_y, log_reg_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(train_x, train_y)

# Predictions
rf_pred = rf.predict(test_x)

# Evaluation
rf_acc = accuracy_score(test_y, rf_pred)
print(f"Random Forest Accuracy: {rf_acc:.4f}")
print("\nClassification Report:\n", classification_report(test_y, rf_pred))
print("\nConfusion Matrix:\n", confusion_matrix(test_y, rf_pred))
```

```python
from xgboost import XGBClassifier

# Initialize and train the model
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(train_x, train_y)

# Predictions
xgb_pred = xgb.predict(test_x)

# Evaluation
xgb_acc = accuracy_score(test_y, xgb_pred)
print(f"XGBoost Accuracy: {xgb_acc:.4f}")
print("\nClassification Report:\n", classification_report(test_y, xgb_pred))
print("\nConfusion Matrix:\n", confusion_matrix(test_y, xgb_pred))
```
